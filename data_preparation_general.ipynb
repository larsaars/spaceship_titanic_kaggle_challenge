```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('./datasets/original/train.csv')
```

```python
df
```

```python
df.describe()
```

```python
df['Transported'].sum() / len(df)
```

```python
# make passenger group an extra feature
df['PassengerGroup'] = df['PassengerId'].apply(lambda x: x.split('_')[0])
```

```python
len(df['PassengerGroup'].unique())
```

```python
# make cabin deck / num /side extra features 
df['CabinDeck'] = df.loc[~df['Cabin'].isnull(), 'Cabin'].apply(lambda x: x.split('/')[0])
df['CabinNum'] = df.loc[~df['Cabin'].isnull(), 'Cabin'].apply(lambda x: x.split('/')[1])
df['CabinSide'] = df.loc[~df['Cabin'].isnull(), 'Cabin'].apply(lambda x: x.split('/')[2])
```

```python
# delete cabin feature
del df['Cabin']
```

```python
# make last name a feature
df['LastName'] = df.loc[~df['Name'].isnull(), 'Name'].apply(lambda x: x.split()[-1])
```

```python
# encode labels to numeric (keep NaN values)
mask = df.isnull()
df = df.apply(LabelEncoder().fit_transform)
df[mask] = np.nan
```

```python
df.describe()
```

```python
# write to csv
df.to_csv('./datasets/preprocessed/train.csv', index=False)
```
